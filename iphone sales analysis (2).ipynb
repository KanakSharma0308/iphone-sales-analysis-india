#iphone sales analysis in india

```python
most_rated=df.sort_values(by = ["Star Rating"],ascending=False)
top10=most_rated.head(10)
print(top10)
```

```python
#lets have a look at the number of ratings of the highest rated iphone on flipkart
```

```python
import plotly.io as pio
pio.renderers.default="iframe"
iphone=top10["Product Name"].value_counts()
labels=iphone.index
counts=top10["Number Of Ratings"]
figure=px.bar(top10,x=labels,y=counts,title="number of highest rated iphone")
figure.show()
#print(\"labels is=\",labels)
#print(\"count)
```

```python
iphone=top10["Product Name"].value_counts()
labels=iphone.index
counts=top10["Number Of Reviews"]
figure=px.bar(top10,x=labels,y=counts,title="number of highest rated iphone")
figure.show()
```
